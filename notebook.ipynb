## Introduction

We are the analytical team from a Bitcoin trading company, we want the know the popularity of trend regarding the trading bitcoin within 10 years. Plus, we will demonstrate the returns of bitcoin within recent 10 years in terms of different time periods of investment for the potential investors. Furthermore, we would like the generate some insights for investors from the historical performance of view, which may guide them to invest wisely in the future. Therefore, in this analysis report, we are working the historical data of bitcoin base on the following key questions:
1. Is the bitcoin investment getting popular within 10 years?
2. What is the price trend of bitcoin from 2012-01 to 2021-03? Total return? Annual return? Monthly return?
3. Is the bitcoin categorised as high volatility or risk investment for the potential individual investors base on the historical performance?
4. Is there any insights and recommendations of bitcoin investment for the future investors?


## Set up

In first step, we download the Bitcoin Historical Data- bitstampUSD 1-minute data from 2012-01-01 to 2021-03-31 from Kaggle which include timestamp, open, high, low, close, Volume_Currency_, Volume_BTC_and Weighted_Price columns. 
([Bitcoin dataset link](http://www.kaggle.com/datasets/mczielinski/bitcoin-historical-data))

## Import

Import this dataset and table into Big query cloud system.
